```python
import os
from glob import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import geopandas as gpd
from shapely.geometry import Point
import earthpy as et
import seaborn as sns
```

```python
sns.set(font_scale=1.5)
sns.set_style("white")

data = et.data.get_data('spatial-vector-lidar')
os.chdir(os.path.join(et.io.HOME, 'earth-analytics'))
```

```python
blder_df = np.array([[476911.31,  4429455.35]])
geometry = [Point(xy) for xy in blder_df]

blder_loc = gpd.GeoDataFrame(geometry,
                            columns=['geometry'],
                            crs={'init': 'epsg:2957'})
```

```python
fig, ax = plt.subplots(figsize=(10,10))
blder_loc.plot(ax=ax)
ax.set_title("Plot of Boulder, CO Location")

plt.ticklabel_format(useOffset=False)
plt.show()
```

```python
blder_df_geog = blder_loc.to_crs(epsg=4326)
blder_df_geog.crs
```
